```r
# Loading required libraries
library('ggplot2')
```

```r
# Loading the ISLR gene expression dataset (Ch10Ex11.csv)
filename = "C:/Users/hamma/Desktop/Newcastle University/Data Science/Machine Learning/c0070933_hammad_mir_extended_technical_project/dataset/skin cancer/HAM10000_metadata.csv"
metadata = read.csv(filename, header=TRUE)
dim(metadata)
```

```r
# Checking the number of samples and features
head(metadata)
```

```r
# Metadata summary
summary(metadata$age)
```

```r
ggplot(data.frame(table(metadata$age)), aes(x=Var1, y=Freq, fill=Var1)) +
geom_bar(stat="identity") +
geom_text(aes(label=Freq), vjust=1.6, color="black", size=3.5)+
theme(legend.position = 'none', , plot.title=element_text(hjust=0.5)) +
labs(title = "Age distribution plot", x = "Age", y = "Frequency")
```

```r
# locations of the leisons
(loc <- data.frame(table(metadata$localization)))

ggplot(loc, aes(x=Var1, y=Freq, fill=Var1)) +
geom_bar(stat="identity") +
geom_text(aes(label=Freq), vjust=1.6, color="black", size=3.5)+
theme(axis.text.x = element_text(angle = 20, hjust = 1), legend.position="none", plot.title=element_text(hjust=0.5)) +
labs(title = "Leision localization frequency plot", x = "Leision Location", y = "Frequency")
```
